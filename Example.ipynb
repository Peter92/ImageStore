```python
import requests
from PIL import Image

from io import BytesIO
from ImageStore import DecodeImage, EncodeImage
```

```python
testfile = "test.png"
test = Image.open(testfile)
test
```

```python
encoded = EncodeImage("Heck", False)
encoded.set_background(testfile)
encoded.save("encoded.png");
```

```python
Image.open("encoded.png")
```

```python
decoded = DecodeImage("encoded.png")
print(decoded)
```

```python
img_url = "https://i.imgur.com/VcHO7h6.png"

r = requests.get(img_url)
if r.status_code == 200:
    img = Image.open(BytesIO(r.content))
img
```

```python
encoded = EncodeImage("I'm only 30 though...", False)
encoded.set_background(img_url)
encoded.save("encoded2.png");
```

```python
Image.open("encoded2.png")
```

```python
decoded = DecodeImage("encoded2.png")
print(decoded)
```
